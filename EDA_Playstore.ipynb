```python
# convert the whole size column into bytes

#let's define a function

def convert_size(size):
  if isinstance(size, str):
    if 'k' in size:
      return float(size.replace('k',"")) * 1024
    elif 'M' in size:
      return float(size.replace('M',"")) * 1024  * 1024
    elif 'Varies with device' in size:
      return np.nan
  return size
```

```python
df['Size'] = df['Size'].apply(convert_size)
```

```python
df['Size']
```

```python
# rename the column
df.rename(columns={'Size' :'Size_in_bytes'},inplace=True)
```

```python
df.head()
```

```python
df['Size_in_bytes'] = pd.to_numeric(df['Size_in_bytes'], errors='coerce')
df['Size_in_Mb']=df['Size_in_bytes'].apply(lambda x :x /(1024*1024))
```

```python
df['Size_in_kb']=df['Size_in_bytes'].apply(lambda x :x /(1024))
```

## Installs column
1. Remove + sign

2. remove ,

3. convert the column into an integer

```python
df['Installs'] =df['Installs'].apply(lambda x :x.replace('+', "") if '+' in str(x) else(x))
```

```python
df.head()
```

```python
df['Installs']=df['Installs'].apply(lambda x:x.replace(',', "")if ','in str(x)  else (x) )
```

```python
df['Installs']
```

```python
df['Installs'] = df['Installs'].replace('Free', np.nan)
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
df['Installs'] = df['Installs'].fillna(0).astype(int)
```
