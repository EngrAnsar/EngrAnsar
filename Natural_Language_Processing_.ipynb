```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

text = "Education is a powerful tool that helps individuals develop knowledge, skills, and values. It plays a vital role in shaping personality and improving communication abilities such as reading, writing, speaking, and listening. Education also creates better job opportunities and helps people contribute positively to society. Through education, individuals can improve their standard of living and work towards reducing poverty. Overall, education is essential for personal growth and social development."

stop_words = set(stopwords.words('english'))

# Tokenize and filter stopwords
tokens = word_tokenize(text)
filtered_words = [word for word in tokens if word.lower() not in stop_words]

print("Filtered Words:", filtered_words)
```

```python
from nltk import pos_tag, ne_chunk
from nltk.tokenize import word_tokenize

text = "Barack Obama was the 44th president of the United States."

# Tokenize and tag
tokens = word_tokenize(text)
pos_tags = pos_tag(tokens)

# Named entity recognition
entities = ne_chunk(pos_tags)
print("Named Entities:", entities)
```

```python
from nltk.stem import PorterStemmer
import nltk
nltk.download('wordnet')

text = "running runs runner"

# Stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in text.split()]
print("Stemmed Words:", stemmed_words)
```
